```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
telcodata = pd.read_csv('telcodata.csv')
telcodata.head()
totalcharge = pd.DataFrame(telcodata['StreamingMovies'])  # Response
tenure = pd.DataFrame(telcodata['tenure'])      # Predictor
streamingtv = pd.DataFrame(telcodata['StreamingTV'])
TechSupport = pd.DataFrame(telcodata['TechSupport'])  
DeviceProtection = pd.DataFrame(telcodata['DeviceProtection'])  
OnlineBackup = pd.DataFrame(telcodata['OnlineBackup'])  
OnlineSecurity = pd.DataFrame(telcodata['OnlineSecurity'])  
PhoneService  = pd.DataFrame(telcodata['PhoneService']) 
MultipleLines  = pd.DataFrame(telcodata['MultipleLines']) 
Dependents  = pd.DataFrame(telcodata['Dependents']) 
Partner  = pd.DataFrame(telcodata['Partner']) 
SeniorCitizen  = pd.DataFrame(telcodata['SeniorCitizen']) 
Contract  = pd.DataFrame(telcodata['Contract']) 
PaperlessBilling  = pd.DataFrame(telcodata['PaperlessBilling']) 
PaymentMethod  = pd.DataFrame(telcodata['PaymentMethod']) 
Churn  = pd.DataFrame(telcodata['Churn']) 
trainDF = pd.concat([tenure, totalcharge], axis = 1).reindex(tenure.index)
trainDFstreamingtv = pd.concat([tenure, streamingtv], axis = 1).reindex(tenure.index)
trainDFTechSupport = pd.concat([tenure, TechSupport], axis = 1).reindex(tenure.index)
trainDFDeviceProtection = pd.concat([tenure, DeviceProtection], axis = 1).reindex(tenure.index)
trainDFOnlineBackup = pd.concat([tenure, OnlineBackup], axis = 1).reindex(tenure.index)
trainDFOnlineSecurity = pd.concat([tenure, OnlineSecurity], axis = 1).reindex(tenure.index)
trainDFInternetService = pd.concat([tenure, InternetService], axis = 1).reindex(tenure.index)
trainDFMultipleLines = pd.concat([tenure, MultipleLines], axis = 1).reindex(tenure.index)
trainDFPhoneService = pd.concat([tenure, PhoneService], axis = 1).reindex(tenure.index)
trainDFDependents = pd.concat([tenure, Dependents], axis = 1).reindex(tenure.index)
trainDFPartner = pd.concat([tenure, Partner], axis = 1).reindex(tenure.index)
trainDFSeniorCitizen = pd.concat([tenure, SeniorCitizen], axis = 1).reindex(tenure.index)
trainDFContract = pd.concat([tenure, Contract], axis = 1).reindex(tenure.index)
trainDFPaperlessBilling = pd.concat([tenure, PaperlessBilling], axis = 1).reindex(tenure.index)
trainDFPaymentMethod = pd.concat([tenure, PaymentMethod], axis = 1).reindex(tenure.index)
trainDFChurn = pd.concat([tenure, Churn], axis = 1).reindex(tenure.index)
trainDFInternetService = pd.concat([tenure, InternetService], axis = 1).reindex(tenure.index)
trainDFInternetService = pd.concat([tenure, InternetService], axis = 1).reindex(tenure.index)
```
